```python
import numpy as np
import scipy.signal as signal
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# Load the audio file
audio_path = 'first_collection/xuzhaoqi.wav'
y, sr = librosa.load(audio_path, sr=None)
```

```python
# Apply a Butterworth filter for noise reduction
def butterworth_filter(data, cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    y_filtered = signal.filtfilt(b, a, data)
    return y_filtered
```

```python
# Define the cutoff frequency and order of the filter
cutoff_frequency = 4000  # in Hz
filter_order = 5
```

```python
# Apply the filter
y_filtered = butterworth_filter(y, cutoff_frequency, sr, filter_order)
```

```python
# Plot the original and filtered signals
plt.figure(figsize=(15, 6))
plt.subplot(2, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Original Audio')
plt.subplot(2, 1, 2)
librosa.display.waveshow(y_filtered, sr=sr)
plt.title('Filtered Audio')
plt.tight_layout()
plt.show()
```

```python
# Save the filtered audio to a new file
# librosa.output.write_wav('first_collection/filtered_xuzhaoqi.wav', y_filtered, sr)
import soundfile as sf
# 使用soundfile库保存滤波后的音频
sf.write('first_collection/filtered_xuzhaoqi.wav', y_filtered, sr)
print("The audio has been successfully filtered and saved as 'filtered_xuzhaoqi.wav'")
```
